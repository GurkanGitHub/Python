```python
# Je souhaite appliquer la fonction racine à la variable population, quelle fonction dois-je utiliser ?

df = pd.DataFrame([["Belgium", 11.3, 30510, "Brussels"], 
                   ["France", 64.3, 671308, "Paris"], 
                   ["Germany", 81.3, 357050, "Berlin"], 
                   ["Netherlands", 16.9, 41526, "Amsterdam"], 
                   ["United Kingdom", 64.9, 244820, "London"]],  
                  index= range(5), 
                  columns = ["country", "population", "area", "capital"])
print(df, "\n")

print(np.sqrt(df.population), "\n")
df.population.apply(np.sqrt)
```

```python
# Quelle est la commande qui permet uniquement d'avoir la distribution de la variable area ?

df.population.describe()
```

```python
# Sélectionner uniquement la capital et la population des pays ayant une densité supérieur à 200

df = df.set_index("country")
```

```python
df["density"] = [370.37, 95.782, 27.69, 406.97, 265.09]
df["density_ratio"] = [1.35, 0.35, 0.83, 1.48, 0.97]
```
