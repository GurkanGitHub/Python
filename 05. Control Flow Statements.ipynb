#### Basic structure of these statements

```python
if condition1:
    execute body1
elif condition2:
    execute body2
else:
    execute body3
```

#### How does the structure of statement work?

```python
‚Ä¢ condition1 is checked first,
    ‚óã if the result is True the body1 in the 'if' statement' will be executed. 
    ‚óã If not, condition2 in the 'elif' statement is checked. 
        If the result is True, body2 will be executed, 
        if not, body3 in the last control point 'else' will be executed.
```

## 4. Nested 'if-elif-else' Statements

Both if-else and if-elif statements can be nested. Let's see the nested structure on the same movie ticket example.

```python
audience_group = 'kid', 'teen', 'adult'
audience = "teen"
if audience in audience_group:
    if audience == "kid":
        print("it is free to go to cinema")
    elif audience == "teen":
        print("discounted price!")
    else: # audience == "adult":
        print("normal price")
else:
    print("No such audience, stay at your home!")
```

### Fizz Buzz Problemi:

```python
# 1'den 50 ye kadar olan sayƒ±larƒ± ekrana yazdƒ±ran bir program yazalƒ±m. 
    # Ancak programƒ±mƒ±z, 3'e b√∂l√ºnen rakamlar yerine "Fizz", 
    # 5'e b√∂l√ºnebilen sayƒ±lar yerine "Buzz", 
    # her iklsine de b√∂l√ºnebilenlere ise "FizzBuzz" yazacak. 
# Programƒ± Jupyter Notebook'ta ve a≈üaƒüƒ±da belirtilen 3 farklƒ± ≈üekilde yazalƒ±m.
# For d√∂ng√ºs√º kullanƒ±larak,
# While d√∂ng√ºs√º ile,
# List comprehension ile
```

### with for loop

```python
for i in list(range(1,51)):
    if (i % 3 == 0) & (i % 5 == 0):
        print("fizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz") 
    else: print(i)
```

### with while loop

```python
i=1
while i<51:
    if i %3 ==0 and i%5 == 0:
        print("fizzbuzz")
    elif i %3 == 0:
        print("fizz")
    elif i %5 == 0:
        print("buzz")  
    else:
        print(i)
    i+=1
```

### with list comprehension

```python
["fizzBuzz" if (i % 3 == 0) and (i % 5 == 0) else "Fizz" if i % 3 == 0 else "Buzz" if i % 5 == 0 else i for i in range(1,51)]
```
