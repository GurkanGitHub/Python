### Divide dataset into target and independent variable

```python
inputs = df.drop("salary_more_then_100k", axis = 1)
target = df["salary_more_then_100k"]
```

```python
inputs
```

```python
target
```

### Converting categorical columns into number

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# We need 3 different objects
le_company= LabelEncoder()
le_job= LabelEncoder()
le_degree= LabelEncoder()
```

```python
inputs["company_n"] = le_company.fit_transform(inputs["company"])
inputs["job_n"] = le_job.fit_transform(inputs["job"])
inputs["degree_n"] = le_degree.fit_transform(inputs["degree"])
```

```python
inputs
```

```python
inputs =inputs.drop(["company", "job", "degree"], axis = 1)
inputs
```

### Train the classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Model Setup
dtm = DecisionTreeClassifier(criterion='gini')
```

$$Gini Impurity value = 1-p(yes)^{2}- p(no)^{2}$$

```python
1-(0/3)**2-(3/3)**2
```

```python
#Training the model
dtm.fit(inputs, target)
```

```python
#predicting score
dtm.score(inputs, target)
```

```python
dtm.predict([[2,2,1]]) #google, sales exec, masters
```

### Classify with entropy

$$Surprise = log(\frac 1{probability})$$

$$Entropy = \sum log(\frac 1{p(x)}*p(x))$$

```python
#Model Setup
dtm = DecisionTreeClassifier(criterion='entropy')
#Training the model
dtm.fit(inputs, target)
#predicting score
dtm.score(inputs, target)
```

### Predict survival of the passenger 

```python
df = pd.read_csv("C:/Users/Owner/py/ML/9_decision_tree/Exercise/titanic.csv")
```

```python
df.head()
```

```python
df = df.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis=1)
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Sex_n= LabelEncoder()
```

```python
df["Sex_n"] = Sex_n.fit_transform(df["Sex"])
df = df.drop("Sex", axis = 1)
df.head()
```

```python
df.groupby("Pclass").mean()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:return 24
    else:return Age
```

```python
df["Age"] = df[["Age", "Pclass"]].apply(impute_age, axis=1)
```

```python
df.isnull().sum()
```

```python
df["Age"].value_counts()
```

```python
X = df.drop("Survived", axis = 1)
y = df["Survived"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

### Classifier make up 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtm = DecisionTreeClassifier()
```

### Training the model

```python
dtm.fit(X_train, y_train)
```

```python
y_pred = dtm.predict(X_test)
```

```python
y_pred = dtm.predict(X_test)
```

```python
#predicting score
dtm.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cnf_mtrx = confusion_matrix(y_test, y_pred )
```

```python
cnf_mtrx 
```
