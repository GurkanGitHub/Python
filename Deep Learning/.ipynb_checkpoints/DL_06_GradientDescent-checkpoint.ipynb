```python
df = pd.read_csv("insurance_data.csv")
```

**Split train and test set**

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["age", "affordibility"]]
y = df.bought_insurance
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 25)
```

**Preprocessing: Scale the data so that both age and affordibility are in same scaling range**

```python
#scaling: bring age and affordibility to the same scale
X_train_scaled = X_train.copy()
X_train_scaled["age"] = X_train_scaled["age"]/100
```

```python
X_test_scaled = X_test.copy()
X_test_scaled["age"] = X_test_scaled["age"]/100
```

**Model Building: First build a model in keras/tensorflow and see what weights and bias values it comes up with. We will than try to reproduce same weights and bias in our plain python implementation of gradient descent. Below is the architecture of our simple neural network**

<img src = "nn.png" height=800 width=800/>

```python
import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, # with only one layer, neuron
                       input_shape = (2,), # age and affordibility are 2 params in the input
                       activation = "sigmoid",
                       kernel_initializer="ones", #weights starts with 1
                       bias_initializer = "zeros" # bias starts with 0
                       )
                          ])


model.compile(optimizer="adam",
              loss = "binary_crossentropy",
              metrics = ["accuracy"]
              )

model.fit(                          # model fit runs the training.
          X_train_scaled, y_train,
          epochs = 500 # tra,l and error: initialy 1000 and up, 5000 gives the best accuracy
          )
```

**Evaluate the model on test set**

```python
model.evaluate(X_test_scaled,y_test)
```

```python
model.predict(X_test_scaled)#anything that bigger than 0.5, it will buy the insurance
```

**Now we start implementing gradient descent in plain python. Again the goal is to come up with same w1, w2 and bias that keras model calculated. We want to show how keras/tensorflow would have computed these values internally using gradient descent**

**First write couple of helper routines such as sigmoid and log_loss**

```python
#implementing gradient descend function in python from scratch
```

```python
# sigmoid vector function, takes vector, not one value
def sigmoid_numpy(X):
  return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
import numpy as np
def log_loss(y_test, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i, epsilon)for i in y_predicted]
  y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_test*np.log(y_predicted_new) + (1-y_test)*np.log(1-y_predicted_new))
```

**All right now comes the time to implement our final gradient descent function !! yay !!!**

```python
#defining gradient descent function: helps to find weights, w1, w2, bias

def gradient_descent(age, affordibility, y_true, epochs, loss_threshold):
  w1 = w2= 1
  bias = 0
  rate = 0.5 # see if it runs fast
  n = len(age)

  for i in range(epochs):
    weighted_sum = w1 * age + w2 * affordibility + bias
    y_predicted = sigmoid_numpy(weighted_sum)

    loss = log_loss(y_true, y_predicted)

    # now operate the weights based on derivative function

    w1d = (1/2)*np.dot(np.transpose(age), (y_predicted - y_true))
    w2d = (1/2)*np.dot(np.transpose(affordibility), (y_predicted - y_true))

    bias_d = np.mean(y_predicted - y_true)

    w1 = w1 - rate * w1d
    w2 = w2 - rate * w2d
    bias = bias-rate *bias_d

    print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

    if loss <= loss_threshold:
      break
  return w1, w2, bias

```

```python
gradient_descent(X_train_scaled["age"], X_train_scaled["affordibility"], y_train, 1000, 0.6252)
```
