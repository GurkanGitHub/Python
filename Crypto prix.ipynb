```python
#define a few variables to stitch together our desired GET request. 
```

```python
base_url = "https://api.gemini.com"
```

```python
api_version = "v2"
```

```python
symbol = "btcusd"
```

```python
time_frame="1day" 
```

```python
url = f"{base_url}/{api_version}/candles/{symbol}/{time_frame}"
```

```python
url
```

```python
#sending request
```

```python
response = requests.get(url=url)
```

```python
response
```

```python
response.json()
```

```python
# get the raw response JSON data into a more analysis-friendly format
```

```python
def get_candles_dataframe(json_data:str)->pd.DataFrame:
    candles_df = pd.DataFrame(json_data)
    candles_df.columns = ["time", "open_price", "high_price", "low_price", "close_price", "volume"]
    candles_df["time"] = candles_df["time"].apply(lambda x: datetime.datetime.fromtimestamp(x/1000.0))
    return candles_df
```

```python
btcusd_prices_df = get_candles_dataframe(response.json())
```
