```python
import pandas as pd
```

```python
hc = pd.read_excel("D:/barr/Bloomberg.xlsx")
```

```python
hc.head()
```

```python
hc.columns
```

```python
import datetime as dt
```

```python
hc["month"] = hc["Unnamed: 0"].dt.month_name()
hc["day"] = hc["Unnamed: 0"].dt.day_name()
```

```python
hc.head()
```

```python
hc = hc.loc[hc["month"].isin(["February", "March", "April", "May"])]
```

```python
hc.head()
```

```python
hcF = hc[hc["month"]=="February"]
```

```python
hcF
```

```python
hcF = hcF.replace("x", 2)
```

```python
hcF
```

```python
hcF= hcF.groupby(by="day").sum()
```

```python
hcF
```

```python
hcF = hcF.T
```

```python
hcF
```

```python
hcF
```

```python
hcF= hcF.reindex(['2am', '4am', '11am', '1pm', '4pm', '6pm', '8pm', '10pm'])
```

```python
hcF
```

```python
hcF.columns
```

```python
hcF = hcF[['Monday','Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]
```

```python
hcF.rename({"Monday": "Mon_hc", "Tuesday": "Tue_hc", "Wednesday": "Wed_hc", "Thursday": "Thu_hc", 
                 "Friday": "Fri_hc", "Saturday": "Sat_hc", "Sunday": "Sun_hc"}, axis=1, inplace=True)
hcF
```

```python
hcF.index
```

```python
hcF.rename(index={'2am':2, '4am':4, '11am':11, '1pm':13, '4pm':16, '6pm':18, '8pm':20, '10pm':22}, inplace =True)
```

```python
hcF
```

```python
res_feb = pd.read_excel("D:/barr/February.xlsx")
```

```python
res_feb
```

```python
res_feb.drop([0,1,3,5,6,7,8,9,10,12,14,15,17,19,21,23], inplace=True, axis=0)
```

```python
res_feb.rename({"Hour of Day": "day"}, inplace=True, axis=1)
```

```python
res_feb.set_index("day", inplace=True)
```

```python
res_feb.rename({"Monday": "Mon_res","Tuesday": "Tue_res", "Wednesday": "Wed_res", 
                  "Thursday": "Thu_res", "Friday": "Fri_res", "Saturday": "Sat_res", "Sunday": "Sun_res" }, axis = 1, inplace=True)
```

```python
res_feb
```

```python
february = pd.concat([hcF,res_feb], axis=1)
```

```python
february
```

```python
february.columns
```

```python
february = february[['Mon_hc','Mon_res', 'Tue_hc','Tue_res', 'Wed_hc', 'Wed_res', 'Thu_hc', 'Thu_res', 
               'Fri_hc', 'Fri_res', 'Sat_hc', 'Sat_res', 'Sun_hc','Sun_res']]
```

```python
february
```

```python
means = []
for i in february:
    means.append(february[i].mean())
means
```

```python
february.loc[len(february)] = means
```

```python
february = february.round(1)
```

```python
february
```

```python
february.rename(index = {8:"Mean"}, inplace=True)
february
```

```python
february.to_excel("february2023_comparision.xlsx")
```
